```python
import numpy as np
import pandas as pd


Prueba_A = np.random.normal(loc=7.2, scale=0.4, size=52)
Prueba_B = np.random.gamma(shape=306, scale=0.0229, size=52)
```

```python
df2 = pd.DataFrame({
    "A": Prueba_A,
    "B": Prueba_B
    })
```

```python
df2.describe()
```

Descripcion:
segun el count hay 52 elemento, seguen el mean la media de A es de 7.2 y la media de B es de 6.9, la desviacion estandar de A es de  0.4 y de la B es 0.3, el minimo de A es de 6.13, y el minimo de B es de 6.09, en el cuantil 25% de A es de 6.8 y de B es de 6.7, el cuantil de A es de 7.17 y el de B es de 6.9, el cuantil de 75% es de 7.60 y el de B es 7.17 y el maximo de A es de 8.27 y el de B es de 7.5.

```python
from scipy import stats
```

```python
from scipy.stats import shapiro
shapiro([Prueba_A])
```

```python
from scipy.stats import shapiro
shapiro([Prueba_B])
```

# En ambas pruebas  si se cumple el supuesto de normalidad con una confianza del 95%

```python
stats.ttest_rel(Prueba_A, Prueba_B)
```

# si se rechaza que el puntaje medio en ambas encuestas es igual con una confianza del 95%

```python
stats.kendalltau(Prueba_A, Prueba_B)

```

# No se rechaza con una confianza del 95% que los puntajes de ambas encuestas no están correlacionados

```python
t_stats, p_val = stats.spearmanr(df2["A"], df2["B"])
print("Estadístico t:", t_stats)
print("Valor p:", p_val)
```
